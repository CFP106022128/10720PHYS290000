```python
import numpy as np
import matplotlib.pyplot as plt
import math


#SI unit
h=6.62*10**-34
c=3*10**8
k=1.39*10**-23

num=100

v=np.linspace(10**0,10**15.5,100,endpoint=True)

#T=float(input("Input temperature"))
T=5000
B=2*h*v**3/c**2/(math.e**(h*v/k/T)-1)

#plt.xlim(10**11,10**15.2)
#plt.ylim(10**-14,10**-6)
plt.plot(v,B,".")
#plt.loglog(v,B,".")


plt.xlim(0,10**15)

dB=np.random.normal(0.,10**-8.5,num)
B += dB

plt.plot(v,B)
#plt.loglog(v,B)

plt.show()
```
