```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.random(1000)
print(x)
```

```python
plt.hist(x,100)

plt.xlabel("x value")
plt.ylabel("y Number of elements")
plt.show()
```

```python
mu, sigma =0, 0.3 #mean and standard deviation
s = np.random.normal(mu,sigma,10) #the last number represents the number of sample
plt.hist(s,10)

#plt.xlim(-2,2)
plt.show()
```

```python
ss=plt.hist(s,10)  #It shows the true data of the plot
print(ss)


```

```python
bins=np.linspace(-0.5,0.5,21,endpoint=True)
print(bins)
plt.hist(s,bins)

#plt.xlim(-2,2)
plt.show()
```
